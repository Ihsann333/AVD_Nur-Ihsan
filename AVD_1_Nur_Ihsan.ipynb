# **Business understanding**

**Business Objective**
Tujuan utama proyek ini yaitu untuk mengidentifikasi pola harga dan spesifikasi mobil sport yang paling diminati oleh berbagai segmen pelanggan untuk meningkatkan efektivitas pemasaran dan rekomendasi produk bagi dealer atau platform e-commerce otomotif.

**Asses Situation**
Situasi bisnis yang mendasari analisis ini adalah situasi dari dealer dan platform e-commerce otomotif yang sering kesulitan memahami pola permintaan mobil sport, menentukan harga optimal, dan menargetkan pelanggan dengan rekomendasi yang sesuai. Kurangnya wawasan berbasis data dapat menyebabkan harga yang tidak kompetitif dan strategi pemasaran yang kurang efektif.

**Data Mining Goals**
Mengidentifikasi pola harga mobil sport berdasarkan variabel seperti merek, model, tahun produksi, dan spesifikasi teknis (horsepower, engine size, transmission, dll.). Dengan analisis ini, pelanggan dapat dikelompokkan berdasarkan preferensi harga dan spesifikasi,untuk membantu menetapkan harga yang kompetitif, sementara sistem rekomendasi berdasarkan data akan menyarankan mobil sport yang sesuai dengan minat pelanggan.

**Project Plan**
Proyek ini akan dimulai dengan memeriksa dan membersihkan data agar siap digunakan. Setelah itu, analisis pola harga akan dilakukan berdasarkan merek, model, tahun, dan spesifikasi mobil sport. Kemudian, pelanggan akan dikelompokkan berdasarkan preferensi mereka agar dealer bisa lebih tepat dalam menawarkan mobil. Selanjutnya, model prediksi harga akan dibuat untuk membantu menentukan harga yang sesuai dengan pasar. Sistem rekomendasi juga akan dikembangkan agar pelanggan bisa mendapatkan saran mobil yang cocok. Terakhir, semua hasil akan dievaluasi dan digunakan untuk membantu strategi pemasaran dan penjualan yang lebih efektif.
